```python
import pandas as pd

df = pd.read_csv("orders.csv")

df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

numeric_cols = ["list_price", "discount_percent", "cost_price"]
for col in numeric_cols:
    if col not in df.columns:
        df[col] = 0 
    df[col] = pd.to_numeric(df[col]) 

# Trim spaces from all string columns
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)

# Calculate Discount Amount, Sale Price, and Profit
df["discount"] = df["list_price"] * df["discount_percent"] / 100
df["sale_price"] = df["list_price"] - df["discount"]
df["profit"] = df["sale_price"] - df["cost_price"]

# Calculate totals
total_discount = df["discount"].sum()
total_sale_price = df["sale_price"].sum()
total_profit = df["profit"].sum()

total_row = pd.DataFrame([{
    "list_price": df["list_price"].sum(),
    "discount_percent": None,  
    "cost_price": df["cost_price"].sum(),
    "discount": total_discount,
    "sale_price": total_sale_price,
    "profit": total_profit
}])

df = pd.concat([df, total_row], ignore_index=True)

display(df)  

```

```python
#13.Find the percentage of revenue contributed by each category

cursor.execute("""
    SELECT 
        p.category,
        ROUND(SUM(p.sale_price * p.quantity), 2) AS category_revenue,
        ROUND((SUM(p.sale_price * p.quantity) / 
              (SELECT SUM(sale_price * quantity) FROM Products)) * 100, 2) AS revenue_percentage
    FROM Products p
    JOIN Orders o ON p.order_id = o.order_id
    GROUP BY p.category
    ORDER BY revenue_percentage DESC
""")

order_detail_table = cursor.fetchall()

Query13 = pd.DataFrame(order_detail_table, columns=["category", "category_revenue", "revenue_percentage"])

display(Query13)
```

```python
#17. Product Performance Query

cursor.execute('''
 SELECT 
        p.product_id,
        p.category,
        p.sub_category,
        SUM(p.quantity) AS total_quantity_sold,
        ROUND(SUM(p.sale_price * p.quantity), 2) AS total_revenue,
        ROUND(SUM(p.profit), 2) AS total_profit,
        ROUND(AVG(p.discount_percent), 2) AS avg_discount_percent,
        RANK() OVER (ORDER BY SUM(p.sale_price * p.quantity) DESC) AS revenue_rank
    FROM Products p
    JOIN Orders o ON p.order_id = o.order_id
    GROUP BY p.product_id, p.category, p.sub_category
    LIMIT 10
''')

ranked_products = cursor.fetchall()

Query17 = pd.DataFrame(ranked_products, columns= [ "product_id", "category", "sub_category", 
    "total_quantity_sold", "total_revenue", "total_profit", "avg_discount_percent", "revenue_rank"])

display(Query17)

```

```python
#18. Identify products with discounts greater than 20% and calculate the impact of discounts on sales

cursor.execute("""
    SELECT 
        p.product_id,
        o.order_id,
        o.region,
        p.category,
        p.sub_category,
        p.discount_percent,
        p.sale_price,
        p.quantity,
        p.discount,
        (p.discount * p.quantity) AS total_discount_amount,
        (p.sale_price * p.quantity) AS total_sales_after_discount,
        ((p.list_price * p.quantity) - (p.sale_price * p.quantity)) AS discount_impact
    FROM Products p
    JOIN Orders o ON o.order_id = p.order_id
    WHERE p.discount_percent > 0.20
    ORDER BY p.discount_percent DESC
    LIMIT 10
""")

result18 = cursor.fetchall()

Query18 = pd.DataFrame(result18, columns=[
    "product_id", "order_id", "region", "category", "sub_category",
    "discount_percent", "sale_price", "quantity", "discount",
    "total_discount_amount", "total_sales_after_discount", "discount_impact"
])

display(Query18)

```

```python
#20. Identify products that had the highest order volume and analyze their revenue impact

cursor.execute('''
    SELECT 
        p.product_id,
        p.category,
        p.sub_category,
        SUM(p.quantity) AS total_order_volume,
        ROUND(SUM(p.sale_price * p.quantity), 2) AS total_revenue,
        ROUND(SUM(p.profit), 2) AS total_profit,
        RANK() OVER (ORDER BY SUM(p.quantity) DESC) AS volume_rank,
        RANK() OVER (ORDER BY SUM(p.sale_price * p.quantity) DESC) AS revenue_rank
    FROM Products p
    JOIN Orders o ON p.order_id = o.order_id
    GROUP BY p.product_id, p.category, p.sub_category
    LIMIT 10;
''')

order_detail_table = cursor.fetchall()

Query20 = pd.DataFrame(order_detail_table, columns=['product_id', 'category', 'sub_category', 'total_order_volume',
    'total_revenue', 'total_profit', 'volume_rank', 'revenue_rank'])

display(Query20)
```

```python
import json
import re

with open("project1.ipynb", "r", encoding="utf-8") as f:
    notebook_data = json.load(f)

queries = []

for cell in notebook_data.get("cells", []):
    if cell.get("cell_type") == "code":
        source_lines = cell.get("source", [])
        code = "\n".join(source_lines).strip()

      # Extract lines between triple quotes if cursor.execute(''' ... ''') is used
        sql_blocks = re.findall(r"cursor\.execute\s*\(\s*(?:r?'''|r?\"\"\")([\s\S]*?)(?:'''|\"\"\")\s*\)", code)

        if sql_blocks:
            for sql in sql_blocks:
                cleaned_sql = sql.strip()
                if "SELECT" in cleaned_sql.upper():
                    queries.append(cleaned_sql)
        else:
            
            if "SELECT" in code.upper():
                queries.append(code)


with open("queries.json", "w", encoding="utf-8") as json_file:
    json.dump(queries, json_file, indent=4)

print("✅ Clean SQL queries saved to queries.json")
```
